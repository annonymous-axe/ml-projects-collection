```python
# Postive setiment - Joy, Love, Surprise
# Negative sentiment - Sadness, Anger, Fear

df['label'].replace(to_replace='sadness', value=1, inplace=True)
df['label'].replace(to_replace='anger', value=1, inplace=True)
df['label'].replace(to_replace='fear', value=1, inplace=True)
df['label'].replace(to_replace='joy', value=0, inplace=True)
df['label'].replace(to_replace='love', value=0, inplace=True)
df['label'].replace(to_replace='surprise', value=0, inplace=True)
```

```python
lm = WordNetLemmatizer()
```

```python
def transoformation(df_column):
    output = []
    for i in df_column:
        new_text = re.sub('[^a-zA-Z]', ' ', str(i))
        new_text = new_text.lower()
        new_text = new_text.split()
        new_text = [lm.lemmatize(j) for j in new_text if j not in set(stopwords.words('english'))]  
        output.append(' '.join(str(k) for k in new_text))
    return output
```

```python
var = transoformation(df.text)
```

```python
def sentimental_analysis(input_):
    new_input = transoformation(input_)
    transformed_input = cv.transform(new_input)
    prediction = rfc.predict(transformed_input)
    if prediction[0] == 0:
        print("Negative Sentiment")
    elif prediction[0] == 1:
        print("Postive Sentiment")
    else:
        print("Invalid Sentiment")
```
